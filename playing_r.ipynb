```r
y <- c(rexp(n=100, 1), rexp(n=100, 5), rexp(n=100, 2))
hist(y[1:100])
plot(y)
```

```r
cost_func <- function(y, ts_start, ts_stop){
  sumstart <- y[ts_start: ts_stop]
  mam <- 1/mean(sumstart)
  netsum(-length(sumstart) * mam + mam * sum(sumstart))
}
```
